```python
#Top 10 Offense Groups
offense_group_vals = data.OFFENSE_CODE_GROUP.value_counts()[:10]

display(offense_group_vals / data.shape[0])

Result = (offense_group_vals / data.shape[0])
print(f"shape de Result est {Result.shape}, et de offense_group_vals est: {offense_group_vals.shape}")


plt.figure(figsize=[8,4])
plt.bar(offense_group_vals.index, Result)
plt.title('Top 10 Offense Groups (as % of all crimes)')
plt.xticks(rotation=90) 
plt.show()
```

```python
group_by_year =data.groupby('YEAR')['YEAR'].count()

plt.figure(figsize=[10,6])
x = np.arange(len(group_by_year))
plt.bar(x, group_by_year.values)
plt.title("Number of incident in every year", fontsize=15)
plt.xlabel("Year")
plt.ylabel("Number of incident")
plt.xticks(x,group_by_year.index)
plt.show()
```

```python
OFFENSE_CODE_GROUP_variable = data.groupby('OFFENSE_CODE_GROUP')['OFFENSE_CODE_GROUP'].count().sort_values()

plt.figure(figsize=(18,6))

x = np.arange(len(OFFENSE_CODE_GROUP_variable))

plt.bar(x, OFFENSE_CODE_GROUP_variable.values, edgecolor =  "black", lw=0.8, width=0.7)
plt.xticks(x, OFFENSE_CODE_GROUP_variable.index, rotation =90)
plt.title("Number of OFFENSE in the 3 years")
plt.xlabel("OFFENSE_CODE_GROUP")
plt.ylabel("Number of OFFENSE GROUPE")
plt.show()
```

```python
part_of_data = data.sample(n=30)
STREET = part_of_data.groupby('STREET')['STREET'].count()

value = STREET.values
index = STREET.index

plt.figure(figsize=(12,8))
plt.pie(value,labels = index,autopct="%.2f%%", pctdistance=0.8)
plt.show()
```

unstack() convertit le Series en un DataFrame où les jours de la semaine deviennent des colonnes, facilitant la visualisation des données en format large.


```python
avg_crime = data.groupby(['YEAR', 'MONTH']).count()['INCIDENT_NUMBER'].mean()
year_and_month = data.groupby(['MONTH', 'YEAR']).count()['INCIDENT_NUMBER'].unstack()
```

```python
def style_negative(v, props=''):
    return props if v < avg_crime else None #SO will be return style color:blue if condition verified


s2 = year_and_month.style.applymap(style_negative, props='color:blue;')\
              .applymap(lambda v: 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None)
s2
```
